# 绘制探测中心制作的雷达拼图

```python
%matplotlib inline
import os
from metradar.io.read_new_mosaic_func import decode_mosaic
from metradar.graph.draw_latlon_func import draw_latlon
```

```python
filepath = '/mnt/e/metradar_test/cmadaas/mosaic_cref/2025/07/02/RADA_L3_MST_CREF_QC'

filename = 'Z_RADA_C_BABJ_20250702064819_P_DOR_ACHN_CREF_20250702_064200.bin'
outpath = '/mnt/e/metradar_test/pic'
outname = filename + '.png'
data = decode_mosaic(filepath,filename)

```

```python
slat = 25
nlat = 42
wlon = 100.
elon = 120.

newdata = data.CREF.sel(lat=slice(slat,nlat),lon=slice(wlon,elon))
tstr = filename.split('_')[2]+filename.split('_')[3]

draw_latlon(newdata.data,newdata.lat.data,newdata.lon.data,slat,nlat,wlon,elon,outpath,outname,tstr,
            add_title=1,badd_gis=False,
            prefix_title='雷达组合反射率拼图',units='dBZ',thred=10,dpi=300)
```
